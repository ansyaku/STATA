```r
data.norm<-read.delim("festival.dat", header=TRUE)
```

```r
str(data.norm)
```

```r
library(ggplot2)
hist.day1<-ggplot(
    data.norm, aes(day1))+
    geom_density( colour="black", fill="white", size=1)+
    labs(x="Hygene score on day 1", y="Density")
```

```r
hist.day1
```

```r
hist.day1+stat_function(
    fun=dnorm, 
    args=list(mean=mean(data.norm$day1, na.rm=TRUE)), 
    colour="black", size=1
)
```
